```python
# "RNN" : nn.RNN,
# "LSTM"    : nn.LSTM,
# "GRU"     : nn.GRU
#         self.lstm2 = nn.LSTM(h1, h2, batch_first=True)
#         lstm_out2, _ = self.lstm2(lstm_out1)
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.hidden_size = hidden_size
        self.attn = nn.Linear(hidden_size * 2, hidden_size)  # Adjusted linear layer
        self.v = nn.Linear(hidden_size, 1, bias=False)

    def forward(self, hidden, encoder_outputs):
        seq_len = encoder_outputs.size(1)
        hidden = hidden.unsqueeze(1).repeat(1, seq_len, 1)  # Repeat hidden state for each encoder output
        energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim=2)))
        attention_weights = F.softmax(self.v(energy), dim=1)  # Applied linear layer to energy
        context_vector = torch.bmm(attention_weights.transpose(1, 2), encoder_outputs)
        return context_vector, attention_weights

class NEW_Model2(nn.Module):
    def __init__(self, input_size, h1, h2, output_size):
        super(NEW_Model2, self).__init__()
        self.h = h1
        self.lstm1 = nn.LSTM(input_size, h1, batch_first=True)
        self.attention = Attention(h1)  
        self.fc1 = nn.Linear(h1, output_size)
#         self.fc2 = nn.Linear(h2, output_size)

    def forward(self, x):
        lstm_out1, _ = self.lstm1(x)
        attn_context, attn_weights = self.attention(lstm_out1[:, -1, :], lstm_out1)
        y = F.relu(self.fc1(attn_context.squeeze(1)))
#         y = self.fc2(y)
        return y
```
